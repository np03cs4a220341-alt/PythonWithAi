```python
num_cols = ['Age','Height','Weight','BMI','Temperature','Heart Rate','Blood Pressure','Cholesterol']
for col in num_cols:
    df[col].fillna(df[col].median(), inplace=True)
```

```python
cat_cols = ['Gender','Blood Type','Diabetes','Smoking']
for col in cat_cols:
    df[col].fillna(df[col].mode()[0], inplace=True)
```

```python
df = pd.read_csv("/content/drive/MyDrive/AI/Titanic-Dataset.csv")
subset_df = df[['Name', 'Pclass', 'Sex', 'Age', 'Fare', 'Survived']]
first_class = subset_df[subset_df['Pclass'] == 1]

mean_fare = first_class['Fare'].mean()
median_fare = first_class['Fare'].median()
max_fare = first_class['Fare'].max()
min_fare = first_class['Fare'].min()

print("Mean Fare:", mean_fare)
print("Median Fare:", median_fare)
print("Maximum Fare:", max_fare)
print("Minimum Fare:", min_fare)
```

```python
# Check number of null values in Age column
null_age_count = subset_df['Age'].isnull().sum()
print("Number of null values in Age column:", null_age_count)

# Drop rows where Age is null
subset_df_clean = subset_df.dropna(subset=['Age'])

# Verify that nulls are removed
print("Null values in Age after dropping:", subset_df_clean['Age'].isnull().sum())
```

```python
# 1. One-hot encode 'Embarked'
embarked_dummies = pd.get_dummies(df['Embarked'], prefix='Embarked')

# 2. Add to original DataFrame
df = pd.concat([df, embarked_dummies], axis=1)

# 3. Drop original 'Embarked'
df.drop('Embarked', axis=1, inplace=True)

# 4. Preview results
print(df.head())
```

```python
survival_by_sex = df.groupby('Sex')['Survived'].mean()

print("Mean Survival Rate by Gender:")
print(survival_by_sex)
```

```python
import matplotlib.pyplot as plt


# Calculate mean survival rate
mean_survival = df.groupby('Sex')['Survived'].mean()
print(mean_survival)

# Visualization
plt.figure(figsize=(6,4))
mean_survival.plot(kind='bar')
plt.title("Survival Rate by Gender")
plt.xlabel("Gender")
plt.ylabel("Mean Survival Rate")
plt.xticks(rotation=0)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("/content/drive/MyDrive/AI/Titanic-Dataset.csv")

# Group by Embarked and Sex
survival_by_port_gender = df.groupby(['Embarked', 'Sex'])['Survived'].mean()
print(survival_by_port_gender)

# Visualization
survival_by_port_gender.unstack().plot(kind='bar', figsize=(8,5))
plt.title("Survival Rates by Gender and Embarkation Port")
plt.xlabel("Embarkation Port")
plt.ylabel("Survival Rate")
plt.show()
```
